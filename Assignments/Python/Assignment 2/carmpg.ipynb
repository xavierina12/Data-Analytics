```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#1) Load auto-mpg.data into a DataFrame autodf. 
df=pd.read_csv('carmpg.csv')
autodf=pd.DataFrame(df)
autodf.rename(columns={'car name':'carname'}, inplace=True)
autodf
```

```python
#2) Give description of the generated DataFrame autodf.
autodf.describe()
```

```python
#3) Display the first 10 rows of the DataFrame autodf.
autodf.head(10)
```

```python
#4) Find the attributes which have missing values. 
#Handle the missing values using following two ways: 
    #i. Replace the missing values by a value before that. 
   # ii. Remove the rows having missing values from the original dataset 
autodf.isnull().sum()
```

```python
#5) Print the details of the car which gave the maximum mileage.
a=autodf[autodf.mpg==autodf.mpg.max()]
a.carname
```

```python
#6) Find the average displacement of the car given the number of cylinders. 
autodf.groupby(['cylinders'])['displacement'].mean().sort_values(ascending=False)
```

```python
#7) What is the average number of cylinders in a car? 
autodf.groupby(['carname'])['cylinders'].mean().sort_values(ascending=False)
```
