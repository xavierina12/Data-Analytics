```python
from tkinter import *

root = Tk()
root.title('Simple Calculator')

e = Entry(root, width=35, borderwidth=5)
e.grid(row=0, column=0, columnspan=3, padx=10, pady=10)


def Button_Clicker(number):
    new_number =e.get()+str(number)
    e.delete(0, END)
    e.insert(0,new_number)

def Button_Clear():
    e.delete(0, END)

def Button_Add(first_number):
    global f_num
    global math
    math = "addition"
    f_num = first_number
    e.delete(0, END)
    e.insert(0,f_num+"+")

def Button_Subtract(first_number):
    global f_num
    global math
    math = "subtraction"
    f_num = first_number
    e.delete(0, END)
    e.insert(0,f_num+"-")

def Button_Multiply(first_number):
    global f_num
    global math
    math = "multiplication"
    f_num = first_number
    e.delete(0, END)
    e.insert(0,f_num+"x")

def Button_Divide(first_number):
    global f_num
    global math
    math = "division"
    f_num = first_number
    e.delete(0, END)
    e.insert(0,f_num+"/")

def Button_Equal(snum):
    num_1 = f_num
    num2=''
    for i in range(len(snum)):
        if (snum[i]=='+') or (snum[i]=='-') or (snum[i]=='x') or (snum[i]=='/'):
            num2=snum[i+1:]
            break;
        else:
            pass
        
    #check if float or int
    if (num_1.isdigit()) and (num2.isdigit()):
        num_1=int(num_1)
        num2=int(num2)
    else:
        num_1=float(num_1)
        num2=float(num2)   
    
    if math == "addition":
        str1=str((num_1+num2))
        str1=snum+"="+str1
        e.delete(0, END)
        e.insert(0,str1)
        
    if math == "subtraction":
        str1=str((num_1-num2))
        str1=snum+"="+str1
        e.delete(0, END)
        e.insert(0,str1)
                 
    if math =="multiplication":
        str1=str((num_1*num2))
        str1=snum+"="+str1
        e.delete(0, END)
        e.insert(0,str1)
        
    if math == "division":
        str1=str((num_1/num2))
        str1=snum+"="+str1
        e.delete(0, END)
        e.insert(0,str1)


button_1 = Button(root, text="1", padx=40, pady=20, command=lambda: Button_Clicker(1),activebackground="white",activeforeground="black")
button_2 = Button(root, text="2", padx=40, pady=20, command=lambda: Button_Clicker(2),activebackground="white",activeforeground="black")
button_3 = Button(root, text="3", padx=40, pady=20, command=lambda: Button_Clicker(3),activebackground="white",activeforeground="black")
button_4 = Button(root, text="4", padx=40, pady=20, command=lambda: Button_Clicker(4),activebackground="white",activeforeground="black")
button_5 = Button(root, text="5", padx=40, pady=20, command=lambda: Button_Clicker(5),activebackground="white",activeforeground="black")
button_6 = Button(root, text="6", padx=40, pady=20, command=lambda: Button_Clicker(6),activebackground="white",activeforeground="black")
button_7 = Button(root, text="7", padx=40, pady=20, command=lambda: Button_Clicker(7),activebackground="white",activeforeground="black")
button_8 = Button(root, text="8", padx=40, pady=20, command=lambda: Button_Clicker(8),activebackground="white",activeforeground="black")
button_9 = Button(root, text="9", padx=40, pady=20, command=lambda: Button_Clicker(9),activebackground="white",activeforeground="black")
button_0 = Button(root, text="0", padx=40, pady=20, command=lambda: Button_Clicker(0),activebackground="white",activeforeground="black")
button_dec = Button(root, text=".", padx=40, pady=20, command=lambda: Button_Clicker('.'),activebackground="white",activeforeground="black")
button_clear = Button(root, text="Clear", padx=30, pady=20, command=Button_Clear,activebackground="white",activeforeground="black")
button_add = Button(root, text="+", padx=39, pady=20, command=lambda: Button_Add(e.get()))
button_equal = Button(root, text="=", padx=91, pady=20, command=lambda: Button_Equal(e.get()),activebackground="white",activeforeground="black")

button_subtract = Button(root, text="-", padx=41, pady=20, command=lambda: Button_Subtract(e.get()),activebackground="white",activeforeground="black")
button_multiply = Button(root, text="x", padx=41, pady=20, command=lambda: Button_Multiply(e.get()),activebackground="white",activeforeground="black")
button_divide = Button(root, text="/", padx=41, pady=20, command=lambda: Button_Divide(e.get()),activebackground="white",activeforeground="black")

button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)

button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)

button_0.grid(row=4, column=0)
button_dec.grid(row=4, column=1)
button_clear.grid(row=4, column=2, columnspan=3)
button_add.grid(row=5, column=0)
button_equal.grid(row=5, column=1, columnspan=2)

button_subtract.grid(row=6, column=0)
button_multiply.grid(row=6, column=1)
button_divide.grid(row=6, column=2)


root.mainloop()
```
